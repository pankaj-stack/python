```python
import tkinter as tk
from tkinter import messagebox as mb
import random
win=tk.Tk()
win.configure(bg='orange')
l=[]
entry_name_list=[]
# button=tk.Button(win,width=100,text='BILLING SYSTEM',activebackground='navy',bg='navy',fg='yellow',
#                 activeforeground='Yellow',font=400,borderwidth=6,height=5)
# button.grid(row=0,columnspan=6,sticky='W',pady=10)



#                                <---------------------------
def searching_of_bill_no():
    try:
        with open(f'/home/pankaj/Downloads/billing_detail/{billing_no.get()}.txt','r') as f:
            datas=f.read()
            mb.showinfo('BILL REPORT',datas)
#             listbox=tk.Listbox(win,width=32,height=17,borderwidth=5)
# #             for i in range(len(datas)):
#             n=0
#             for i in range(len(datas)):
                
#                 if n<5:
#                     if i=='\n':
#                         n+=1
#                     if n<6:
#                         listbox.insert(i,datas[i])
#                         if n==5:
#                             n=0
                

#         listbox.insert(0,datas)
#             listbox.grid(row=0,column=3)
            f.close()
    except FileNotFoundError:
            mb.showerror('ERROR','!!!  NO RECORD FOUND   !!!')
def customer_detail():
    count=0
    count+=1
    button1=tk.Button(win,width=165,bg='navy',borderwidth=6,activebackground='navy')
    button1.grid(row=3,columnspan=6,sticky='W')
    customer_list=['Customer Name','Contact No.','Bill No.']
    p=0
    p1=1
    # marking_label=tk.Label(win,text='Marking Point',fg='navy').grid(row=4)
    for i in range(3):
        label=tk.Label(button1,text=customer_list[i],bg='navy',fg='white',activebackground='navy')
        label.grid(row=2,column=p,padx=50)
        if i==0:
            customer_name=tk.StringVar()
            entry1=tk.Entry(button1,textvariable=customer_name,borderwidth=5)
            entry1.grid(row=2,column=p1)
            entry_name_list.append(entry1)
        elif i==1:
            customer_mobile=tk.StringVar()
            entry2=tk.Entry(button1,textvariable=customer_mobile,borderwidth=5)
            entry2.grid(row=2,column=p1)
            entry_name_list.append(entry2)
        else:
            search_bill=tk.StringVar()
            entry3=tk.Entry(button1,borderwidth=5,textvariable=search_bill)
            entry3.grid(row=2,column=p1)
            entry_name_list.append(entry3)
        p+=2
        p1+=2 
    button2=tk.Button(button1,width=20,text='SEARCH',background='white',activebackground='navy',command=searching_of_bill_no)
    button2.grid(row=2,column=8,padx=20)
    return customer_name,customer_mobile,search_bill
x,y,billing_no=customer_detail()

            
cosmetics=['Bath Soup','Face  Cream','Face Wash','Hair Spray','Hair Gel','Body Lotion']
grocery=['Rice','Food Oil','Daal','Wheat','Sugar','Tea']
cold_drinks=['Maza','Cock','Frooti','Thumbs UP','Limca','Sprite']

for j in range(3):
    button3=tk.Button(win,text='button3',bg='Navy',activebackground='navy',height=18,width=30,borderwidth=6)
    button3.grid(row=0,column=j,pady=2)
    if 0<=j<=2:
        if j==0:
            item=cosmetics
        elif j==1:
            item=grocery
        elif j==2:
            item=cold_drinks
        for i in range(6):
            ent=4
            label=tk.Label(button3,background='navy',foreground='white',text=item[i])
            label.grid(row=i,column=0,padx=2,pady=10,sticky='W')
            value=tk.StringVar()
            entry='e' + str(ent)
            entry=tk.Entry(button3,textvariable=value,borderwidth=5)
            entry.grid(row=i,column=1)
            entry_name_list.append(entry)
            ent+=1
            l.append(value)
cosmetics_index,grocery_index,cold_drinks_index={},{},{}
def total():
    l1=[10,20,30,40,50,10,20,30,40,50,10,20,30,40,50,13,44,55]
    sum=sum1=sum2=0
    for i in range(len(l)):
        
        if i<6:
            if l[i].get() is not '':
                sum+=int(l[i].get())*l1[i]
                cosmetics_index[cosmetics[i]]=int(l[i].get())*l1[i]
        elif i<12:
            if l[i].get() is not '':
                sum1+=int(l[i].get())*l1[i]
                grocery_index[grocery[i-6]]=int(l[i].get())*l1[i]
        else:
            if l[i].get() is not '':
                sum2+=int(l[i].get())*l1[i]
                cold_drinks_index[cold_drinks[i-12]]=int(l[i].get())*l1[i]
        collection=[sum+(sum*(.00005*100)),sum1+(sum1*(.00005*100)),sum2+(sum2*(.00005*100)),sum*(.00005*100),sum1*(.00005*100),sum2*(.00005*100),]
#         collection=[1,3,5,2,4,6]
        billing_menu_list=['Total Cosmetic Price','Total Grocery Price','Grocery Tax','Cosmetic Tax','Total Cold Drink Price','Cold Drink Tax'] 
        menu_button=tk.Button(win,width=165,height=8,activebackground='navy',bg='navy',borderwidth=6)
        menu_button.grid(row=2,columnspan=6,sticky='W')
        var1,var2=0,0
        en=19
        for i in range(len(billing_menu_list)):
            if 0<=i<=2:
                c=tk.StringVar(value=collection[i])
                label=tk.Label(menu_button,background='navy',fg='white',text=billing_menu_list[i]).grid(row=var1,column=0,sticky='W')
                entry='ee'+str(i)
                entry=tk.Entry(menu_button,textvariable=c,borderwidth=5).grid(row=var1,column=1,padx=10,pady=7)
                entry_name_list.append(entry)
#                 en+=1
                var1+=1
            if i>=3:
                c=tk.StringVar(value=collection[i])
                label=tk.Label(menu_button,background='navy',fg='white',text=billing_menu_list[i]).grid(row=var2,column=2,sticky='W')
                entry='ee'+str(i)  
                entry=tk.Entry(menu_button,textvariable=c,borderwidth=5).grid(row=var2,column=3,padx=10,pady=7)
                entry_name_list.append(entry)
#                 en+=1
                var2+=1
        
    print(sum,sum1,sum2)
list_box_object=[]
def generate_bill():
#         total()
        if x.get()=='' or y.get()=='':
            mb.showerror('Error','PLEASE FILL THE CUSTOMER NAME AND MOBILE NUBMER')
            tk.mainloop()
        else:
            counter=0
            bill_no=random.randint(1000,9001)
            list_box=tk.Listbox(win,width=32,height=17,borderwidth=5)
            list_box.insert(0,f'Bill No.  :       {bill_no}')
            list_box.insert(1,f'Customer Name  :  {x.get()}')
            list_box.insert(2,f'Mobile No.:       {y.get()}')
            for i,v in cosmetics_index.items():
                list_box.insert(counter+3,f'{i}     : {v}')
                counter+=1
            for i,v in grocery_index.items():
                list_box.insert(counter+3,f'{i}     : {v}')
                counter+=1
            for i,v in cold_drinks_index.items():
                list_box.insert(counter+3,f'{i}     : {v}')
                counter+=1
        
            list_box.grid(row=0,column=3)
            list_box_object.append(list_box)
            choice=mb.askquestion('CONFIRAMATION','DO YOU WANT TO SAVE YOUR BILL DETAILS')
            if choice=='yes':
#                 saving_of_bill()
                 with open(f'/home/pankaj/Downloads/billing_detail/{bill_no}.txt','w') as f:
                    f.write(f'Bill No. : {bill_no}\n\n')
                    f.write(f'Customer Name : {x.get()}\n\n')
                    f.write(f'Mobile No.    : {y.get()}\n\n')
                    for i,v in cosmetics_index.items():
                        f.write(f'{i} : {v}\n\n')
                    for i,v in grocery_index.items():
                        f.write(f'{i} : {v}\n\n')
                    for i,v in cold_drinks_index.items():
                        f.write(f'{i} : {v}\n\n')
                    f.close()
            tk.mainloop()
menu_button=tk.Button(win,width=162,height=6,activebackground='navy',bg='navy',borderwidth=6)
menu_button.grid(row=1,columnspan=6,sticky='W')
submit_list=['Total','Generate Bill','Clear','Exit']
def clear_all_entries():
    for i in entry_name_list:
           if i is not None:
                i.delete(0,END)
#     for i in range(list_box_object[0].size()):
#             list_box_object[0].delete(0)
    try:
        list_box_object[0].delete(0,END)
    except IndexError:
#         mb.showinfo('INFO','NOTHIN TO DELELTE')
          pass
    list_box_object.clear()
    for i in range(len(grocery_index)):
        grocery_index.popitem()
    for i in range(len(cold_drinks_index)):
        cold_drinks_index.popitem()
    for i in range(len(cosmetics_index)):
        cosmetics_index.popitem()

def destroy_the_window():
    win.destroy()

for i in range(len(submit_list)):
    if i==0:
        submit_button=tk.Button(menu_button,bg='aqua',activebackground='aqua',borderwidth=3,height=5,width=10,text=submit_list[i],command=total)
        submit_button.grid(row=6,column=i+4,padx=90)
    elif i==1:
        submit_button=tk.Button(menu_button,bg='aqua',activebackground='aqua',borderwidth=3,height=5,width=10,text=submit_list[i],command=generate_bill)
        submit_button.grid(row=6,column=i+4,padx=90)
            
    elif i==2:
        submit_button=tk.Button(menu_button,bg='aqua',activebackground='aqua',borderwidth=3,height=5,width=10,text=submit_list[i],command=clear_all_entries)
        submit_button.grid(row=6,column=i+4,padx=90)
    elif i==3:
        submit_button=tk.Button(menu_button,bg='aqua',activebackground='aqua',borderwidth=3,height=5,width=10,text=submit_list[i],command=destroy_the_window)
        submit_button.grid(row=6,column=i+4,padx=90)
          
            
            
            
            
            
win.geometry(f'{win.winfo_screenwidth()}x{win.winfo_screenheight()}')
win.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

root= tk.Tk() # create window

canvas1 = tk.Canvas(root, width = 300, height = 300)
canvas1.pack()

def ExitApplication():
    MsgBox = tk.messagebox.askquestion ('Exit Application','Are you sure you want to exit the application',icon = 'warning')
    if MsgBox == 'yes':
       root.destroy()
    else:
        tk.messagebox.showinfo('Return','You will now return to the application screen')
        
button1 = tk.Button (root, text='Exit Application',command=ExitApplication)
canvas1.create_window(150, 150, window=button1)
  
root.mainloop()
```

```python
from tkinter import *
from tkinter import filedialog
from tkinter import scrolledtext
import docxpy
import os

def final_search(dir_path,keyword):
    matched_file_list=[]
    all_docx_files=[]
    all_files_list=os.listdir(dir_path)
    for file in all_files_list:
        if(file.endswith(".docx")):  
            all_docx_files.append(file)
            text=docxpy.process(f"{dir_path}/{file}")  
            if(keyword.lower() in text.lower()):
                matched_file_list.append(file)
    return all_files_list,all_docx_files,matched_file_list


#function for browse button
def browse():
    dir_path=filedialog.askdirectory()
    entry_browse.insert(0,dir_path)

def search():
     text_box.place(x=900,y=300)
     dir_path=entry_browse.get()
     keyword=entry_key.get()
     all_files,all_docx,matched_files =final_search(dir_path,keyword)
     text_box.delete('1.0',END)    #to del previous result 1.0= from 0th char of ist line
     res_str=""
     res_str=res_str+f"Total files found:{len(all_files)}\n"
     res_str=res_str+f"Total Docx files found:{len(all_docx)}\n"
     res_str=res_str+f"Total Matched files found:{len(matched_files)}\n"
     for res in matched_files:
           res_str=res_str+"  >"+res+"\n"  #print("\t>",res)

        
     text_box.insert(END,res_str)
    

def reset():
    #print('this is reset button') #prints msg on console
    s=entry_browse.get()
    ln=len(s)
    entry_browse.delete(0,ln) #to clear username text field

    s=entry_key.get()
    ln=len(s)
    entry_key.delete(0,ln)

    text_box.delete('1.0',END)
    

root=Tk()
root.state('zoomed')
root.configure(bg='dark olive green')
root.resizable(width=False,height=False)
root.title("Welcome window")

lbl_title=Label(root,text='WELCOME TO RESUME FINDER',font=('cambria',25,'bold'),bg='dark olive green',fg='white')
lbl_title.pack(pady=10)

lbl_browse=Label(root,text="Directory Path",font=('cambria',22,'italic'),bg="dark olive green",fg="white")
lbl_browse.place(x=280,y=220)

lbl_key=Label(root,text="Enter Keyword",font=('cambria',22,'italic'),bg="dark olive green",fg="white")
lbl_key.place(x=280,y=300)

entry_browse=Entry(root,font=('cambria',20),bg='dark olive green',fg='white',bd=5)
entry_browse.place(x=550,y=220)

entry_key=Entry(root,font=('cambria',20),bg='dark olive green',fg='white',bd=5)
entry_key.place(x=550,y=300)

btn_browse=Button(root,command=browse,text='Browse',font=('Book Antiqua',20),bg='dark olive green',fg='white')
btn_browse.place(x=900,y=220)

btn_srch=Button(root,command=search,text='Search',font=('Book Antiqua',15),bg='dark olive green',fg='white')
btn_srch.place(x=400,y=380)

btn_rst=Button(root,command=reset,text='Reset',font=('Book Antiqua',15),bg='dark olive green',fg='white')
btn_rst.place(x=550,y=380)

text_box=scrolledtext.ScrolledText(root,width=37,height=15,font=('Book Antiqua',15))

root.mainloop()
```
