```python
# INTEGRATED 
from pymysql  import *
def transactions():
    # OCT 27 2019
    ##### Banking Receipt ORIGAINAL 
    from datetime import date
    from datetime import datetime
    connection=connect(host='localhost',user='root',port=3306,password=''
                   ,database='BANKING_SYSTEM2')
    cursor=connection.cursor()
    def receipt():
        query='SELECT * FROM BANKING2'
        cursor.execute(query)
        print('     DATE   TIME       DEPOSIT    WITHDRAW   BALANCE')
        print('_____________________________________________________________')
        for i in cursor.fetchall():
            for j in i:
                if j==0:
                    j=str(j)
                    j=' '
            print(f'{i[0]}   {i[1]}       {i[2]}         {i[3]}' )
            print('__________________________________________________________')
    def deposit_total():
        query='SELECT DEPOSIT FROM BANKING2'
        cursor.execute(query)
        tup1=()
        tot1=0
        for i in cursor:
            tup1+=i
        for j in tup1:
            tot1+=j
        return tot1 
    deposit_total()
    def withdraw_total():
        query='SELECT WITHDRAW FROM BANKING2'
        cursor.execute(query)
        tup1=()
        tot1=0
        for i in cursor:
            tup1+=i
        for j in tup1:
            tot1+=j
        return tot1 
    withdraw_total()

    def deposit():
        now = datetime. now()
        current_time = now. strftime("%H:%M:%S")
        query="INSERT INTO BANKING2(DEPOSIT,DATE,BALANCE) VALUES (%s,%s,%s)"
        deposit_amount=(float(input('enter the deposit amount :')))
        x=deposit_total()+deposit_amount
        y=withdraw_total()
        update=(x-y)
        today=date.today()
        time=today.strftime("%d/%m/%y")
        time=time +'  '+ current_time
        values=(deposit_amount,time,update)
        cursor.execute(query,values)
        return cursor
    def withdraw():
        now = datetime. now()
        current_time = now. strftime("%H:%M:%S")
        query="INSERT INTO BANKING2(WITHDRAW,DATE,BALANCE) VALUES (%s,%s,%s)"
        withd_amount=(float(input('enter the withdraw amount :')))
        update=x-(y+withd_amount)
        today=date.today()
        time=today.strftime("%d/%m/%y")
        time=time +'  '+ current_time
        values=(withd_amount,time,update)
        cursor.execute(query,values)
        return cursor
    
    while True:
        print('1.DEPOSIT        2.WITHDRAW')
        print('3.CHECK BALANCE  4.CHANGE PIN')
        print('5.RECIEPT        6.SEND MONEY')
        print('7.EXIT\n')
    
        option=input('PLEASS ENTER YOUR CHOICE :')
        if option=='1' or option=='2' or option=='3' or option=='4' or option=='5' or option=='6' or option=='7':
            if option=='1':
                deposit()
                x,y=deposit_total(),withdraw_total()
                print(f'balance : {x-y}')
            if option=='2':
                query='SELECT DEPOSIT FROM BANKING2'
                cursor.execute(query)
                withdraw()
                x,y=deposit_total(),withdraw_total()
                print(f'balance : {x-y}')
            if option=='3':
                x,y=deposit_total(),withdraw_total()
                print(f'balance : {x-y}')
            if option=='5':
                receipt()
                print()
            if option=='7':
                print('THANK YOU FOR USEING OUR SERVICES')
                break
        else:
            print(f'!!!!!!!!!! SORRY YOU HAVE ENTERED THE WRONG CHOICE !!!!!!!!!!\n' )
    connection.commit()
    connection.close()
transactions()
```
