tuple packing and unpacking

```python
a = 10
b = 20 
c = 30 
d = 40 

t = a,b,c,d # is called packing
l = [a,b,c,d] # is called packing

print(f't : {t}, {type(t)}')
print(f'l : {l}, {type(l)}')
```

```python
a,b,c,d = t # is called unpacking
aa,bb,cc,dd = l # is called unpacking

print(a,b,c,d,'for object tuple')
print(aa,bb,cc,dd,'for object list')
```

```python
from collections.abc import Hashable
'''
- in searching algorithm hashable object searching is very fast. 
- list is not hashable. 
- tuple is hashable 
- hashable object you can only add into the set or in case of dictionary as a key.

'''
t = (10,20,30)
l = [10,20,30]

print(isinstance(t,Hashable)) # True
print(isinstance(l,Hashable)) # False 
```

```python
hash(l)
```

```python
hash(t)
```
