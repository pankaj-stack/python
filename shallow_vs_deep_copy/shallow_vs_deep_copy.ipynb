```python
l1 = [10,20,30,40] 
l2 = l1 # aliasing ,creating the duplicate references 
print(f'l1 : {l1}')
print(f'l2 : {l2}')
```

```python
l1 = [10,20,30,40] 
l2 = l1 # aliasing ,creating the duplicate references 
l2[2] = 70 #  if we are changing in the l2 then changes will be reflected in both list l1 and l2 
print(f'l1 : {l1}')
print(f'l2 : {l2}')
```

```python
l1 = [10,20,30,40] 
l2 = l1.copy() # creating the duplicate object 
print(f'l1 : {l1}')
print(f'l2 : {l2}')
```

```python
l1 = [10,20,30,40] 
l2 = l1.copy() # creating the duplicate object 
l2[2] = 70 # now ,here changes will be only in l2 
print(f'l1 : {l1}')
print(f'l2 : {l2}')
```

```python
l1 = [10,20,[30,40],50] 
l2 = l1.copy() # creating the duplicate object ,shallow copy
l2[0] = 70 # now ,here changes will be only in l2 
l2[2][0] = 300 # this will reflect in both the the list l1 and l2
print(f'l1 : {l1}')
print(f'l2 : {l2}')
```

```python
import copy 

l1 = [10,20,[30,40],50] 
l2 = copy.copy(l1) # this is called shallow copy ,it not copying completely 

l2[0] = 100 # this change is reflected only in l2 
l2[2][0] = 300  # this change is reflected in both the list l1 and l2 

print(f'l1 : {l1} id of l1 : {id(l1)} , id of l1[2] : {id(l1[2])}')
print(f'l2 : {l2} id of l2 : {id(l2)} , id of l2[2] : {id(l2[2])}')
```

```python
import copy 

l1 = [10,20,[30,40],50] 
l2 = copy.deepcopy(l1) # this is called deepcopy copy ,it copyies completely 

l2[0] = 100 # this change is reflected only in l2 
l2[2][0] = 300  # tthis change is reflected only in l2 

print(f'l1 : {l1} id of l1 : {id(l1)} , id of l1[2] : {id(l1[2])}')
print(f'l2 : {l2} id of l2 : {id(l2)} , id of l2[2] : {id(l2[2])}')
```

<p>
  If the original object does not contain any nested objects, then we should go
  for shallow cloning.
</p>
<p>
  If the original object contains any nested objects, then we should go for deep
  cloning.
</p>
<p>if nesting of objects are not there then .copy, coppy.copy and copy.deepcopy() behaves same like deepcopy but in reality they are shallow copy if you add nested object in that object like list of list  </p>
