- del is a keyword in python responsible for garbage collector. 
- pupose of del keyword : to delete the referece variable or variable. 

```python
x = 10 
print(x)
del x 
# print(x) # <<---------- this will show error because object is delated and object 10 is eligible for garbage collection.
```

```python
s1 = 'durga'

s2 = s1 # when multiple references for a single object then if we delete one of them ojbect then still it will not be eligible for garbage collection until you delete all the references of that object.
s3 = s2  

del s1 
print(s2,s3,sep='\n')
```

```python
# with the help of del keyword we can delete multiple references at a time 
del s2,s3  

# print(s2,s3) # it will show error 
```
