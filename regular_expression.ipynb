```python
# search():
# Syntax: re.search(pattern, string)

# - It is similar to match() but it doesn't restrict us
# to find the matches at the beginning of string only.
# - Returns a Match object if there is a match.
# - If there is more than one match, only the first
# occurrence of the match will be returned.
# - If no matches are found, the value None is returned.
```

```python
msg='start of a sentence and then the sentence end'
match3= re.search('of',msg)
print(match3)

match4= re.search('nce',msg)
print(match4)

match5= re.search('nice',msg)
print(match5)
```

```python
# findall():
# Syntax: re.findall(pattern, string)
# - This method returns a list containing a list of all
# matches of a pattern within the string.
# - It returns the patterns in the order they are found.
# - If there are no matches, then an empty list is returned.
msg='start of a sentence and then the sentence end'
match6= re.findall('nce',msg)
print(match6)
```

```python
# split():
# Syntax: re.split(pattern, string, maxsplit=0)
# - This method returns a list where the string has been
# split at each match.
msg='start of a sentence and then the sentence end'
data1=re.split(' ',msg)
print(data1)

data2= re.split(' ',msg,4)
print(data2)
```

```python
# sub():
# Syntax: re.sub(pattern, replace, string, count=0)
# - It helps to search a pattern and replace with a new
# sub string.
# - If pattern not found, string is returned unchanged.
msg='start of a sentence and then the sentence end'
data3= re.sub(' ','#',msg)
print(data3)

data4= re.sub(' ','_',msg,4)
print(data4)
```

```python
# {} Curly Braces:
# - {m,n} means atleast m and atmost n repetitions of
# the pattern left to it.
# - {m} means exactly m repetitions of the pattern left
# to it.
pattern=re.compile('\d{3}-\d{3}-\d{4}')
match= pattern.finditer(text)
for i in match:
    print(i)
################################
l=re.findall(pattern,text)
for i in l:
    print(i)
```
