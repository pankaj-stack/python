aliasing and cloning

```python
d1 = {1:'a',2:'b'}
d2 = d1 # this is called aliasing,creation duplicate reference variable is called aliasing

d2[500] = 'c' # if you change any one of dict (d1 or d2) then changes will be reflected in both dictionary

print(d1)
print(d2)
```

```python
d1 = {1:'a',2:'b'}
d2 = d1.copy() # this is called cloning,creation duplicate object is called cloning

d2[500] = 'c' # if you change any one of dict (d1 or d2) then changes will be reflected in that dictionary only.

print(d1)
print(d2)
```
