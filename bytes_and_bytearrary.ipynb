- If you want to handle binary data like images, videos and audio files then you should use bytes or bytesarray datatype.
- In case of bytes ,it takes only from 0 - 255 range of values
- bytes are immutable in nature . 

```python
l = [10,20,30,40] 
b = bytes(l) 
print(type(b))
```

```python
for i in b:
    print(i)
```

```python
l1 = [10,20,30,255] 
bytes(l1)
```

```python
l2 = [10,20,30,256] # ValueError: bytes must be in range(0, 256)
# bytes(l2)
```

```python
l = [10,20,30,40]
b = bytes(l)
print(b[0])
print(b[:3])
```

```python
# b[0] = 77 # TypeError: 'bytes' object does not support item assignment
```

----------------------bytearray---------------

- bytes and bytearray me bs itna hi farak hai ki :->
    - bytes     ========> immutable 
    - bytearray ====>     mutable  
    - baaki sari property same hai 

```python
l = [10,20,30,40]
ba = bytearray(l)
```

```python
l = [10,20,30,256] # ValueError: byte must be in range(0, 256)
# ba = bytearray(l)
```

```python
l = [10,20,30,255] # ValueError: byte must be in range(0, 256)
ba = bytearray(l)
for i in ba:
    print(i)
```

```python
print(ba[0])
print(ba[2])
print(ba[-1])
print(ba[1:])
```

```python
ba[0] = 77 
for i in ba: 
    print(i)
```
